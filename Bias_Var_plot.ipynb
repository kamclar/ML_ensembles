```python
#https://towardsdatascience.com/mse-and-bias-variance-decomposition-77449dd2ff55
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

# Dummy dataframe

df_proxy = pd.DataFrame(data={'RMSE': [10, 20, 30, 40],
                         'stds':  [1, 2, 3, 4]}, index=['model_1', 'model_2', 'model_3', 'model_4'])

df_proxy['MSE'] = df_proxy.RMSE.map(np.square)
df_proxy['vars'] = df_proxy.stds.map(np.square)
df_proxy['bias'] = np.array(df_proxy.MSE) - np.array(df_proxy.vars)
df_proxy = df_proxy.T


# PLOTTING

labels = df_proxy.columns

fig, axs = plt.subplots(ncols=3, figsize=[20,5])
fig.suptitle('MODELS bias variance decomposition')


for i, item in enumerate(df_proxy.index[2:]):
    sns.lineplot(ax=axs[i], x=labels, y=df_proxy.loc[item]).set_title(item)
    axs[i].tick_params('x', labelrotation=45) 
```
